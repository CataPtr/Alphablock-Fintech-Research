```python
import pandas as pd
import numpy as np
```

```python
#Read all data and concatenate in a single DF

df1 = pd.read_csv ('S&P 500 2000-2004.csv')
df2 = pd.read_csv ('S&P 500 2004-2009.csv')
df3 = pd.read_csv ('S&P 500 2009-2014.csv')
df4 = pd.read_csv ('S&P 500 2014-2018.csv')
df5 = pd.read_csv ('S&P 500 2018-2023.csv')

pdList = [df1, df2, df3, df4, df5]

df = pd.concat(pdList)
df.reset_index(drop=True, inplace=True)
df
```

```python
#convert data types, sort data and remove unused columns

df['Date'] = pd.to_datetime(df['Date'])

df=df.sort_values(['Symbol Guid','Date'],ascending = [True,True])
df.reset_index(drop=True, inplace=True)

df=df.drop(['R1','R2','R3','R4'],axis='columns')

df.R5.astype('float')
```

```python
#convert dataframe into a numpy array

matx=np.array(df)
matx
```

```python
#create bins

col=[2]
for i in range(len(matx)):
    for j in col:
        if matx[i,j]<=10:
            matx[i,j]=1
        elif matx[i,j]<=20 and matx[i,j]>10:
            matx[i,j]=2    
        elif matx[i,j]<=30 and matx[i,j]>20:
            matx[i,j]=3
        elif matx[i,j]<=40 and matx[i,j]>30:
            matx[i,j]=4
        elif matx[i,j]<=50 and matx[i,j]>40:
            matx[i,j]=5
        elif matx[i,j]<=60 and matx[i,j]>50:
            matx[i,j]=6
        elif matx[i,j]<=70 and matx[i,j]>60:
            matx[i,j]=7
        elif matx[i,j]<=80 and matx[i,j]>70:
            matx[i,j]=8
        elif matx[i,j]<=90 and matx[i,j]>80:
            matx[i,j]=9
        elif matx[i,j]<=100 and matx[i,j]>90:
            matx[i,j]=10

print(matx)
```

```python
#convert data form np array to dataframe

dd=pd.DataFrame(matx, columns=['Date','Symbol Guid','Bin'])
dd
```

```python
#remove data above bin 10 (benchmark has values above 100)

dd.reset_index(drop=True)
dd= dd[dd['Bin']<=10]
```

```python
#convert data types, sort data and reset index (first column)

dd['Date'] = pd.to_datetime(dd['Date'])
dd=dd.sort_values(['Symbol Guid','Date'],ascending = [True,True])
dd.reset_index(drop=True, inplace=True)
```

```python
#convert data into a numpy array

df=np.array(dd)
df
```

```python
#calculate days spent in each bin

j=1
k=0;

df_zeros=df.shape[0]
final_array=np.zeros(shape=(len(df),4),dtype=object)

for i in range(1,df_zeros):
    if (df[i-1,1]==df[i,1] and df[i-1,2]==df[i,2]):
        j=j+1
    else:
        final_array[k,0]=df[i-1,0]
        final_array[k,1]=df[i-1,1]
        final_array[k,2]=df[i-1,2]
        final_array[k,3]=j
        k=k+1
        j=1
print(final_array)
```

```python
#remove zeros

final_data=final_array[~np.all(final_array == 0, axis=1)]
final_data
```

```python
#export data into CSV file
pd.DataFrame(final_data).to_csv('Final-data-with-bins-days-R5.csv',index=False,header=False)
```
